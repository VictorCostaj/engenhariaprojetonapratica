```python
import duckdb
import pandas as pd
import os 
from datetime import datetime
```

```python
con = duckdb.connect(database='dados_duckdb.db', read_only=False)
```

```python
# fazendo a leitura dos CSV

arquivo = 'z0019_1.csv'
data_ingestao = datetime.now()
df = pd.read_csv(f'../landing/{arquivo}', sep=';')
df['nome_arquivo'] = arquivo
df['data_ingestao'] = data_ingestao #colunas adicionais
df.head()
```

# criando a  tabela Bronze

```python
con.execute("""
    CREATE TABLE IF NOT EXISTS bronze_produtos (
            NATBR VARCHAR, 
            MAKTX VARCHAR, 
            WERKS VARCHAR, 
            MANIS VARCHAR, 
            LABST VARCHAR, 
            nome_arquivo VARCHAR, 
            data_ingestao TIMESTAMP,
        )
""")
```

## Ingest√£o dos dados na tabela que criamos 

```python
con.execute("DROP TABLE IF EXISTS bronze_produtos")
```

```python
con.execute("INSERT INTO bronze_produtos SELECT * FROM df")
```
